Replicate 1, Estimate Pi using monte carlo

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt

N = 10000
M = 0
XCircle=[]
YCircle=[]
XSquare=[]
YSquare=[]

for p in range(N):
    x=random.random()
    y=random.random()
    # Corrected: if statement on a new line and indented
    if (x**2 + y**2 <= 1):
        M+=1
        XCircle.append(x)
        YCircle.append(y)
    else:
        # Corrected: typo 'appe' to 'append', and added 'x' for XSquare
        XSquare.append(x)
        YSquare.append(y)

Pi = 4*M/N
print("N=%d M=%d Pi=%.2f" %(N,M,Pi))

# Corrected: Removed extraneous output from print statement
XLin = np.linspace(0,1)
YLin = []
for x in XLin:
    # Corrected: Removed `plt.grid` from this line, assuming it was a typo/merge
    YLin.append(math.sqrt(1-x**2))

# Corrected: Removed unicode spaces from plt.axis and plt.grid
plt.axis("equal")
# Corrected: Combined/fixed the duplicated and malformed grid calls
plt.grid(which="major")
plt.plot(XLin , YLin, color="red" , linewidth="4")
plt.scatter(XCircle, YCircle, color="yellow",
            marker=".")
# Corrected: Removed unicode spaces and fixed typo 'markerl' to 'marker'
plt.scatter(XSquare, YSquare, color="blue",
            marker=".")
plt.title("Monte Carlo method for Pi estimation")
plt.show()
```

2. Using monte Carlo in Central Limit Theorem

```python
import random
import numpy as np
import matplotlib.pyplot as plt

a=1
b=100
N=10000

# Generate population data from a uniform distribution
DataPop=list(np.random.uniform(a,b,N))

# Plotting the population distribution
plt.hist(DataPop, density=True, histtype='stepfilled', alpha=0.2)
plt.title("Population Distribution (Uniform)")
plt.show()

SamplesMeans = []
num_samples_per_extraction = 100
num_extractions = 1000

for i in range(num_extractions):
    DataExtracted = random.sample(DataPop, k=num_samples_per_extraction)
    DataExtractedMean = np.mean(DataExtracted)
    SamplesMeans.append(DataExtractedMean)

# After collecting all sample means, plot their distribution
plt.figure()
plt.hist(SamplesMeans, density=True, histtype='stepfilled', alpha=0.2)
plt.title("Distribution of Sample Means (Central Limit Theorem)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

3. Use gamma distributidistribution in no. 2

```python
import random
import numpy as np
import matplotlib.pyplot as plt

# Parameters for Gamma Distribution
shape = 2  # k (alpha)
scale = 2  # theta (beta)
N = 10000 # Population size

# Generate population data from a Gamma Distribution
DataPop = list(np.random.gamma(shape, scale, N))

# Plotting the population distribution
plt.hist(DataPop, bins=50, density=True, histtype='stepfilled', alpha=0.2, color='blue')
plt.title(f"Population Distribution (Gamma: Shape={shape}, Scale={scale})")
plt.show()

SamplesMeans = []
num_samples_per_extraction = 100
num_extractions = 1000

for i in range(num_extractions):
    DataExtracted = random.sample(DataPop, k=num_samples_per_extraction)
    DataExtractedMean = np.mean(DataExtracted)
    SamplesMeans.append(DataExtractedMean)

# After collecting all sample means, plot their distribution
plt.figure()
plt.hist(SamplesMeans, bins=30, density=True, histtype='stepfilled', alpha=0.2, color='green')
plt.title("Distribution of Sample Means (Central Limit Theorem for Gamma Distribution)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```
