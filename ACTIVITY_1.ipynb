```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of 100,000 values using a normal distribution
mean = 3
std_dev = 1.2
normal_array = np.random.normal(loc=mean, scale=std_dev, size=100000)

# Define the histogram range and bin width
hist_range = (-0.6, 0.6)
bin_width = 0.05
bins = np.arange(hist_range[0], hist_range[1] + bin_width, bin_width)

# Create a histogram of the values
plt.hist(normal_array, bins=bins, range=hist_range, edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Normal Distribution Values')
plt.xlim(hist_range)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two arrays of 100,000 random integers between 0 and 20
a = np.random.randint(0, 21, size=100000)
b = np.random.randint(0, 21, size=100000)

# Create a new array c where c[i] = a[i] - b[i-1]
# The array c will have a length of 99,999
c = a[1:] - b[:-1]

# Create a histogram of the values in c with a range of -20 to +21
plt.hist(c, bins=range(-20, 23), edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of c values')
plt.xlim([-20, 21])
plt.show()
```

```python
# Create an array of 100,000 values using a normal distribution
mean = 0.3
std_dev = 1.2
normal_array = np.random.normal(loc=mean, scale=std_dev, size=100000)

# Print the average and standard deviation of the array
print(f"Average of the normal array: {np.mean(normal_array)}")
print(f"Standard deviation of the normal array: {np.std(normal_array)}")
```

```python
# Create two arrays of 100,000 random values between -2 and +3.3
random_array_3 = np.random.uniform(low=-2, high=3.3, size=100000)
random_array_4 = np.random.uniform(low=-2, high=3.3, size=100000)

# Create a new array by multiplying the two random arrays
multiplied_array = random_array_3 * random_array_4

# Print the average of the values in the multiplied array
print(f"Average of the multiplied array: {np.mean(multiplied_array)}")
```

```python
# Create two arrays of 100,000 random values between -2 and +3.3
random_array_1 = np.random.uniform(low=-2, high=3.3, size=100000)
random_array_2 = np.random.uniform(low=-2, high=3.3, size=100000)

# Create a new array by summing the two random arrays
summed_array = random_array_1 + random_array_2

# Print the average of the values in the summed array
print(f"Average of the summed array: {np.mean(summed_array)}")
```
